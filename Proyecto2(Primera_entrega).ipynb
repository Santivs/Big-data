```python
### Tasa de desempleo  ###
```

```python
d_unemployment = wb.data.DataFrame('SL.UEM.TOTL.ZS',
                                  economy=["COL","BRA","ARG","CHL","PER","ECU","URY","PRY","VEN","BOL","CAN","MEX","USA","CRI"],
                                  time=range(2002,2020),
                                  labels=True).reset_index(drop=True).set_index('Country').unstack().reset_index()
d_unemployment
```

```python
### PIB percapita en dolares ####
```

```python
d_gdp = wb.data.DataFrame('NY.GDP.PCAP.CD',
                                  economy=["COL","BRA","ARG","CHL","PER","ECU","URY","PRY","VEN","BOL","CAN","MEX","USA","CRI"],
                                  time=range(2002,2020),
                                  labels=True).reset_index(drop=True).set_index('Country').unstack().reset_index()
d_gdp
```

```python
### Acceso a electricidad ###
```

```python
d_electricity = wb.data.DataFrame('EG.ELC.ACCS.ZS',
                                  economy=["COL","BRA","ARG","CHL","PER","ECU","URY","PRY","VEN","BOL","CAN","MEX","USA","CRI"],
                                  time=range(2002,2020),
                                  labels=True).reset_index(drop=True).set_index('Country').unstack().reset_index()
d_electricity
```

```python
 ###  seguridad alimentaria (AREA CULTIVABLE)###
```

```python
d_securitya = wb.data.DataFrame('AG.LND.AGRI.ZS',
                                  economy=["COL","BRA","ARG","CHL","PER","ECU","URY","PRY","VEN","BOL","CAN","MEX","USA","CRI"],
                                  time=range(2002,2020),
                                  labels=True).reset_index(drop=True).set_index('Country').unstack().reset_index()
d_securitya
```

```python
### Tasa Mortalidad neonatal (PORCENTAJE NACIDOS QUE SE MUEREN ANTES DE LOS 28 DIAS###
```

```python
d_morta = wb.data.DataFrame('SH.DYN.NMRT',
                                  economy=["COL","BRA","ARG","CHL","PER","ECU","URY","PRY","VEN","BOL","CAN","MEX","USA","CRI"],
                                  time=range(2002,2020),
                                  labels=True).reset_index(drop=True).set_index('Country').unstack().reset_index()
d_morta
```

```python
datafinal=(d_gdp.merge(d_morta,how='outer', on=['level_0','Country'])
      .merge(d_electricity,how='outer', on=['level_0','Country'])
      .merge(d_unemployment,how='outer', on=['level_0','Country'])
      .merge(d_securitya,how='outer', on=['level_0','Country']))
```

```python
datafinal
```

```python
datafinal.columns=['año', 'pais', 'PIB per capita', 'Tasa Mortalidad', 'Acceso Electricidad', 'Tasa Desempleo', 'Seg Alimentaria']
```
