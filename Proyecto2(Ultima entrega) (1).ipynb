```python
import pandas as pd
import scipy.stats as stats

for year in datafinal['año'].unique():
    print (year)
    datafinal02=datafinal.query('año==@year')
    colombia_data = datafinal02[datafinal02['pais'] == 'Colombia']['Tasa Mortalidad']
    sudamerica_data = datafinal02[datafinal02['pais'].isin(['Argentina', 'Brasil', 'Chile', 'Peru', 'Ecuador', 'Bolivia', 'Uruguay', 'Paraguay', 'Venezuela'])]['Tasa Mortalidad']
    t_statistic, p_value = stats.ttest_1samp(sudamerica_data,colombia_data)
    print("Estadística t:", t_statistic)
    print("Valor p:", p_value)
    alfa = 0.05

    if p_value < alfa:
     print("Se rechaza la hipótesis nula: Hay diferencias significativas entre la tasa de mortalidad en Colombia y Suramérica.")
    else:
      print("No se rechaza la hipótesis nula: No hay diferencias significativas entre la tasa de mortalidad en Colombia y Suramérica.")
```
